```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import io

df = pd.read_csv(io.BytesIO(uploaded['ASIAN-PAINTS.csv']))
df.rename(columns={'Date': 'date', 'Open Price': 'open', 'High Price': 'high', 'Low Price': 'low', 'Close Price': 'close', 'WAP': 'wap', 'No.of Shares' : 'shares', 'No. of Trades' : 'trades', 'Total Turnover (Rs.)' : 'turnover', 'Deliverable Quantity' : 'deliQty', '% Deli. Qty to Traded Qty' : 'deliQtyToTradedQty', 'Spread High-Low' : 'highLowSpread', 'Spread Close-Open' : 'closeOpenSpread'}, inplace = True)
```

```python
df['date'] = pd.to_datetime(df['date'], format = '%d-%B-%Y')
df.set_index('date',inplace = True)
df.sort_index(inplace = True)
```

```python
df.info()
```

```python
plt.figure(figsize=(16,8))
plt.plot(df['close'], label='ASIANPAINTS')

plt.title('Close Price History', fontsize=18)
plt.ylabel('Close Price', fontsize=18)
plt.xlabel('Date', fontsize=18)

plt.legend()
plt.grid(False)

plt.show()
```

```python
def SMA(data, period, column='close'):
    return data[column].rolling(window = period).mean().fillna(method='bfill')
```

```python
df['SMA5'] = SMA(df, 5)
df['SMA10'] = SMA(df, 10)
df['SMA20'] = SMA(df, 20)
```

```python
print(df)
```

```python
# Visualize the close price with SMA also Buy and Sell Signals
plt.figure(figsize=(16,8))
plt.plot(df['close'], alpha=0.5, label='Actual Close Price')
plt.plot(df['SMA5'], alpha=0.5, label='SMA5')
plt.plot(df['SMA10'], alpha=0.5, label='SMA10')
# plt.plot(df['SMA50'], alpha=0.5, label='SMA50')

# Adding text into the visualization
plt.title('Close Price w/ SMA over different time periods', fontsize=18)
plt.ylabel('Close Price', fontsize=18)
plt.xlabel('Date', fontsize=18)

# Give legend
plt.legend()
plt.grid(False)
# Show the graph
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, mean_absolute_error, r2_score
```

```python
SMA5_mse = mean_squared_error(df['close'], df['SMA5'])
SMA10_mse = mean_squared_error(df['close'], df['SMA10'])
SMA20_mse = mean_squared_error(df['close'], df['SMA20'])
print('MSE:',SMA5_mse,SMA10_mse,SMA20_mse)
```

```python
SMA5_rmse = mean_squared_error(df['close'], df['SMA5'], squared = False)
SMA10_rmse = mean_squared_error(df['close'], df['SMA10'], squared = False)
SMA20_rmse = mean_squared_error(df['close'], df['SMA20'], squared = False)
print('RMSE:',SMA5_rmse,SMA10_rmse,SMA20_rmse)
```

```python
SMA5_mape = mean_absolute_percentage_error(df['close'], df['SMA5'])
SMA10_mape = mean_absolute_percentage_error(df['close'], df['SMA10'])
SMA20_mape = mean_absolute_percentage_error(df['close'], df['SMA20'])
print('MAPE:',SMA5_mape,SMA10_mape,SMA20_mape)
```

```python
SMA5_mae = mean_absolute_error(df['close'], df['SMA5'])
SMA10_mae = mean_absolute_error(df['close'], df['SMA10'])
SMA20_mae = mean_absolute_error(df['close'], df['SMA20'])
print('MAE:',SMA5_mae,SMA10_mae,SMA20_mae)
```

```python
SMA5_r2 = r2_score(df['close'], df['SMA5'])
SMA10_r2 = r2_score(df['close'], df['SMA10'])
SMA20_r2 = r2_score(df['close'], df['SMA20'])
print('R2: ', SMA5_r2, SMA10_r2, SMA20_r2)
```
